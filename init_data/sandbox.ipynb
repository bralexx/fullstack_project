```python
db2df_nm = {
  'job_title':'Job Title',
  'job_description':'Job Description',
  'rating':'Rating',
  'estimated_salary':'Salary Estimate',
  'min_salary':'min_salary',
  'max_salary':'max_salary',
  'average_salary':'avg_salary',
  'python_required':'python_yn',
  'excel_required':'excel',
  'r_lang_required':'R_yn',
  'spark_required':'spark',
  'company_size':'Size',
}
df2db_nm = {value:key for key, value in db2df_nm.items()}
```

```python
for key, value in db2df_nm.items():
  if value not in df.columns:
    print(key)
```

```python
def df2db_obj(obj):
  result = {}
  for key in df2db_nm:
    if key in obj.keys():
      result[df2db_nm[key]] = obj[key]
  return result
```

```python
failed_obj = []
failed_r = []
for row_id in tqdm(range(len(df))):
  raw_obj = dict(df.iloc[row_id])
  obj = df2db_obj(raw_obj)
  r = requests.post('http://127.0.0.1:8000/contentlist/post_vacancy/', data=obj)
  if r.status_code != 201:
    failed_obj.append(obj)
    failed_r.append(r)
```

```python
new_failed_obj = []
new_failed_r = []
for obj in tqdm(failed_obj):
  r = requests.post('http://127.0.0.1:8000/contentlist/post_vacancy/', data=obj)
  if r.status_code != 201:
    new_failed_obj.append(obj)
    new_failed_r.append(r)
failed_obj = new_failed_obj
failed_r = new_failed_r
```
